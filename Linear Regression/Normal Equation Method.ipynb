```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("FuelConsumption.csv")
df
```

```python
X=df["ENGINESIZE"]
Y=df["CO2EMISSIONS"]
X=np.array(X)
one=np.ones(len(df))
X_T=np.vstack([one,X])
x=X_T.transpose()
x.shape
# np.linalg.inv()
# X=np.insert(X,0,x,axis=0)
```

```python
X_1=np.dot(X_T,x)
print(X_1)
X_inv=np.linalg.inv(X_1)
print(X_inv)
```

```python
v=np.dot(X_T,Y)
W=np.dot(X_inv,v)
b=W[0]
w=W[1]
```

```python
Y_predicted=np.dot(x,W)
print("Y-predicted",Y_predicted)
error=Y-Y_predicted
SOSE=error.transpose()*error
print("Sum of all Squared Error : ",SOSE.sum())
```
